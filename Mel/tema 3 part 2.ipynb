 # 3.9. Роботска рака дефинирана во сферен координатен систем остварува
произволна позиција во просторот преку следните последователни движења:
транслација за r единици долж z-оската, ротација за агол
 околу y-оската и
ротација за агол околу z-оската. Да се определат нејзините сингуларитети.

```python
r, beta, gamma = sp.symbols('r, beta, gamma')
T = rt.hrotation3('z', gamma) * rt.hrotation3('y', beta) * rt.htranslation3(z=r)
T
```

```python
position = T[:3, 3]
linear_jacobian = position.jacobian([r, beta, gamma])
linear_jacobian
```

```python
det_J = sp.det(linear_jacobian).simplify()
det_J
```

```python
equation = sp.Eq(det_J, 0)
equation
```

```python
interval = sp.Interval(-sp.pi, sp.pi, left_open=True)
interval
```

```python
solution = sp.solveset(det_J, beta, interval)
solution # pri beta=0 ili pi se javuva singularitet
```

```python
interval = sp.Interval(-sp.pi, sp.pi, left_open=True)
interval
```

```python
equation = sp.Eq(det_J, 0)
equation
```

```python
solution = sp.solveset(det_J, theta2, interval)
solution # pri beta=0 ili pi se javuva singularitet
```
